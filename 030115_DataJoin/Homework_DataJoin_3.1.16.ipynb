```python
import agate
```

imported agate

```python
frl13 = agate.Table.from_csv('../frl13.csv')
print(frl13)
print(len(frl13.rows))
```

```python
frl2013 = frl13.where(lambda row: row['FREEREDUCED13'] is not None)
```

created a table for the free and reduced lunch table from 2013, then filtered out the empty cells

```python
frl14 = agate.Table.from_csv('../frl14.csv')
print(frl14)
print(len(frl14.rows))
```

```python
frl2014 = frl14.where(lambda row: row['FREEREDUCED14'] is not None)
```

created one for 2013

```python
frl15 = agate.Table.from_csv('../frl15.csv')
print(frl15)
print(len(frl15.rows))
```

```python
frl2015 = frl15.where(lambda row: row['FREEREDUCED15'] is not None)
```

and 2015

```python
frl1314 = frl2013.join(frl2014, 'CODISTSCH13', 'CODISTSCH14', inner=True)
```

joined the first two tables, merging the county names

```python
print(frl1314)
print(len(frl1314.rows))
```

printed it to make sure I did good

```python
frl131415 = frl1314.join(frl2015, 'COUNTY13', 'COUNTY15', inner=True)
```

merged the new table with frl15

```python
print(frl131415)
print(len(frl131415.rows))
```

printed it to make sure I'd done good once again

```python
change = frl131415.compute([
    ('freereduced_change', agate.PercentChange('FREEREDUCED13', 'FREEREDUCED15'))     
])
```

```python
sorted_change = change.order_by('freereduced_change', reverse=True)
```

```python
for_printing = sorted_change.select(['COUNTY13', 'freereduced_change'])
```

```python
for_printing.print_table()
```
